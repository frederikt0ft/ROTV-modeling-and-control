```python
mu, sigma = 0, 0.4 # mean and standard deviation
noise_frequency = 1
tick_rate = 200
noise_period = tick_rate/noise_frequency 



s_h = 0
s_r = 0
s_p = 0

s_h_list = []
s_r_list = []
s_p_list = []
for x in range(int(50)):
    s_h = np.random.normal(mu, sigma) *      0.5
    s_r = np.random.normal(mu, sigma) *      10  #*3.14/180
    s_p = np.random.normal(mu, sigma) *      5  #*3.14/180
    for x in range(80):
        s_h_list.append(s_h)
        s_r_list.append(s_r)
        s_p_list.append(s_p)
 

tick = np.linspace(0,4000,4000)
print(len(s_h_list))
t = tick/200

fig, axs = plt.subplots(2, 1, constrained_layout=True, gridspec_kw={'height_ratios': [1, 1]})
axs[0].axhline(0, color = "black", linestyle='dotted')
axs[0].plot(t, s_h_list, label = "Heave acc")
axs[0].set_ylabel('Noise acc ($m/s^{2}$)')
axs[0].legend()
axs[0].legend(loc = 'lower left')

axs[1].plot(t, s_r_list,  label = "Roll acc", color = "orange")
axs[1].plot(t, s_p_list,  label = "Pitch acc", color = "g")
axs[1].set_ylabel('Noise acc ($degress/s^{2}$)')
axs[1].legend()
axs[1].legend(loc = 'upper left')

plt.setp(axs[0].get_xticklabels(), visible=False)
plt.show()
```
