```python
import matplotlib.pyplot as plt
import numpy as np

def angle_per_second(pwm):
    if -100 <= pwm <= -17:
        return 0.135 * pwm + 1.5147
    elif -17 <= pwm <= 17:
        return 0
    elif 17 <= pwm <= 100:
        return 0.135 * pwm - 1.5147

# Generate PWM values in the range -100 to 100
pwm_values = np.linspace(-100, 101,20000)

# Compute angle per second for each PWM value
angle_values = [angle_per_second(pwm) for pwm in pwm_values]
plt.style.use('seaborn')
# Plot the function
plt.plot(pwm_values, angle_values,  label = "$\dot{u}_i$")
plt.xlabel('PWM (%)')
plt.ylabel('Angle per second (degrees)')
plt.title('Discontinuous actuator model')
plt.grid(True)
plt.legend(fontsize="12")
plt.savefig("Discontinous_actuator_model.pdf", format="pdf", bbox_inches="tight")

plt.show()

```
